```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from google.colab import files
files.upload()

# Read an image
img = cv2.imread('download(1).jpg')
plt.imshow(img)
plt.title('Original Image')
plt.show()

# Histogram plotting of original image
color = ('b', 'g', 'r')

for i, col in enumerate(color):
    histr = cv2.calcHist([img],[i], None,[256],[0, 256])
    plt.plot(histr, color = col)
    # Limit X - axis to 256
    plt.xlim([0, 256])

plt.title('Histogram of Original Image')
plt.show()

# Histogram Ploting of gray scale image
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_img, cmap=plt.cm.gray)
plt.title('Gray Image')
plt.show()
plt.hist(gray_img.ravel(),256,[0,256])
plt.title('Histogram of Gray Image')
plt.show()

# Histogram Equalization of image
equ = cv2.equalizeHist(gray_img)
plt.imshow(equ, cmap=plt.cm.gray)
plt.title('Equalization Image')
plt.show()

plt.hist(equ.ravel(),256,[0,256])
plt.title('Histogram of Equalization Image')
plt.show()
```
